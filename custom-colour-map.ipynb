```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap, BoundaryNorm
```

```python
# Define discrete colors for up to 12 bands
colors = [
    "blue", "green", "yellow", "red", "cyan", 
    "purple", "orange", "pink", "lime", "brown", 
    "gray", "gold"
]

# Create the colormap
custom_cmap = ListedColormap(colors[:11])  # Use only the required number of bands
```

```python
# Define boundaries for discrete values
bounds = list(range(12))  # Boundaries for 0-11
norm = BoundaryNorm(bounds, custom_cmap.N)

# Generate example data for testing
data = np.random.randint(0, 10, size=(10, 10))
```

```python
# Plot the data
plt.imshow(data, cmap=custom_cmap, norm=norm, aspect='auto')
plt.colorbar(ticks=np.arange(12))  # Add ticks for all bands
plt.title("Discrete Colormap Example")
plt.show()
```
